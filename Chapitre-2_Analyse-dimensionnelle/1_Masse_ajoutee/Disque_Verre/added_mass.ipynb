Added mass computation with finite element analysis
===============




## Theory

To compute the added mass of a vibrating system containing the structure and the fluid around, we need to model at the same time the structure, the fluid and the interaction between the two.

### Structural computation

Let's say that the linearized equations of the structures can be modeled with finite elements by the system :

$$ [M_s]  \{\ddot u_s\}  + [C_s] \{ \dot u_s \} + [K_s] \{u_s \}  = \{F_s\}$$

Where : 
* $[M_s]$, $[C_s]$ and $[K_s]$ are the mass, damping and stifness matrix;
* $\{u_s \}$ are the generalized coordinates for the deplacement : $U(X,t) = \{N_s(X)\}^T \{u_s(t)\} $ with $\{N_s(X)\}$ the structural interpolating functions chosen for computing the finite element analysis;
* $\{F_{s}\}$ are the loads on the structures;

### Fluid computation

To linearized the fluid system, let's say that the fluid is at rest (the same development could be generalized with pertubation around mean flow) and a small perturbation is generated around this resting flow. The pressure generated by this perturbed flow is the acoustic pressure and it's the main variable of the fluid equations for the finite element analysis.

Let's say that the linearized equations of the fluids can be modeled with finite elements by the system :

$$ [M_f]  \{\ddot p_a\}  + [C_f] \{ \dot p_a \} + [K_f] \{p_a \}  = \{F_f\} $$

Where : 
* $[M_f]$, $[C_f]$ and $[K_f]$ are the mass, damping and stifness matrix;
* $\{p_s \}$ are the generalized coordinates for the acoustic pressure : $P_a(X,t) = \{N_f(X)\}^T \{p_a(t)\} $ with $\{N_f(X)\}$ the fluid interpolating functions chosen for computing the finite element analysis;
* $\{F_{f}\}$ are the loads on the fluid;


### Interaction fluid structure

The interaction between the fluid and the structure are described by 2 mutual terms : 
* The action of the fluid on the structure : the loads generated by the acoustic pressure on the fluid $F_{f\rightarrow s} = [R] \{p_a\}$
* The action of the structure on the fluid : each fluid particle must follow the fluid deformation, it's the non-penetration condition which appears in the boundary conditions of the fluid and appears on the loads after transformation as :  $F_{s\rightarrow f} = - \rho_0 [R]^T \{ \ddot u_s \}$

Where $[R]$ represents the interaction matrix between the fluid and the structure, it contains the spatial correlations between the fluid and structural interpolating function on the contact surface $\Gamma$ between fluid and solid : 
$$ R_{i,j}  = \int_{\Gamma} N_{f,j} N_{s,i}.d \vec S$$

Finally the global system of the full system fluid/structure can be represented as :


$$ \begin{bmatrix}
		[M_s]  & 0 \\
		\rho_0 [R]^T & M_f
		\end{bmatrix}  
        \begin{bmatrix}
		\{\ddot u_s\}  \\
		\{\ddot p_a\} 
		\end{bmatrix} + 
        \begin{bmatrix}
		[C_s]  & 0 \\
		0 & [C_f]
		\end{bmatrix}  
        \begin{bmatrix}
		\{\dot u_s\}  \\
		\{\dot p_a\} 
		\end{bmatrix} + 
        \begin{bmatrix}
		[K_s]  & -[R] \\
		0 & [K_f]
		\end{bmatrix}  
        \begin{bmatrix}
		\{ u_s\}  \\
		\{ p_a\} 
		\end{bmatrix}
        = 
		\begin{bmatrix}
		\{F_s\}  \\
		\{F_f\} 
		\end{bmatrix}
        $$
        
### Computing of the added mass

The concept of the added mass is a simplification from a global fluid/structural system to an only structural system by adding an added mass.

So, we'll identifies added mass term to analyse from the global previous system. Let's do some simplifications.

#### Incompressibility

The compressibility of the fluid is included into the fluid mass matrix and the damping matrix : 

$$ [M_f] =  \int_{\Omega_f} \frac{1}{C^2}\{N_f\} \{N_f\}^T dV$$
$$ [C_f] =  \int_{\Omega_f} \frac{4 \mu}{3 \rho_0 C^2}\{\nabla N_f\} \{\nabla N_f\}^T dV$$

Where $C$ is the acoustic velocity in the fluid. If the fluid is incompressible : $C \rightarrow \infty $, and $[M_f]\{\ddot p_a\}$ and $[C_f]\{\dot p_a\}$ become neglibeable against the other terms.

#### Inviscid

The viscosity of the fluid is included into the damping matrix :

$$ [C_f] =  \int_{\Omega_f} \frac{4 \mu}{3 \rho_0 C^2}\{\nabla N_f\} \{\nabla N_f\}^T dV$$

Where $\mu$ is the dynamic viscosity of the fluid. If we considers the fluid as inviscid : $\mu \rightarrow 0 $, and $[C_f]\{\dot p_a\}$ become neglibeable against the other terms.

#### Simplified equations of the fluid

By neglegting, the mass and the damping terms, we obtains the equation of the fluid :


$$ [K_f] \{p_a \} + \rho_0 [R]^T \{\ddot u_s\} = \{F_f\} $$

Then, 

$$ \{p_a \} = - [K_f]^{-1} (\rho_0 [R]^T \{\ddot u_s\} + \{F_f\}) $$

#### Simplified equation of the structure
$$ [M_s]  \{\ddot u_s\}  + [C_s] \{ \dot u_s \} + [K_s] \{u_s \}  = \{F_s\} + [R] \{p_a\}$$

And with the simplifications of the fluid equations, 


$$ [M_s]  \{\ddot u_s\}  + [C_s] \{ \dot u_s \} + [K_s] \{u_s \}  = \{F_s\} - [R][K_f]^{-1} (\rho_0 [R]^T \{\ddot u_s\} + \{F_f\})$$

And then we can identify the added mass matrix $[M_a]$:

$$ ([M_s] + \underbrace{\rho_0[R][K_f]^{-1} [R]^T}_{[M_a]} )  \{\ddot u_s\}  + [C_s] \{ \dot u_s \} + [K_s] \{u_s \}  = \{F_s\} - [R][K_f]^{-1}\{F_f\}$$

The full coupled fluid/structural system is now modeled as an uncoupled structural system with an added mass and an added fluid load. The problem with this expression of added mass is that it's necessary to inverse and multiply very big matrixs, it's numerically very difficult. 

So as for this type of acoustic/structural model we would like to study the vibration of the system, we can focus on calculating the added mass for each modes independently.

###  Modal analysis

For the modal analysis we don't take the forces into account, we considere also that there is no structural damping to work on the natural modes :

$$ ([M_s] + \underbrace{\rho_0[R][K_f]^{-1} [R]^T}_{[M_a]} )  \{\ddot u_s\}   + [K_s] \{u_s \}  = 0$$

considering the vibrating mode : 

$\{ u_s\} = \{U\} \exp(j\omega t) $

We then obtain the generalized eigenvalue problem :

$$   [K_s] \{U \}  =  \omega ^ 2 ([M_s] + [M_a] )\{U \} $$

Isolating the added mass term :

$$   [M_a] \{U \}  =   (\frac{1}{\omega ^ 2}[K_s] - [M_s])\{U \} $$


Now we consider the solutions of the natural eigenvalue-problem without added-mass :

$$   [K_s] \{U_0 \}  =  \omega_0 ^ 2 [M_s]\{U_0 \} $$

And we suppose that there are only the eigenvalues that change with the added mass but the mode basis keep the same form : $\{U \} \simeq \{U_0 \}$

It's now possible to compute the added mass of each modes with the expression :

So for each modes :

$$   m_{a,i}   =   (\frac{1}{\omega_i ^ 2}k_{s,i} - m_{s,i}) $$

with $m_{s,i}$ and $k_{s,i}$ the structural mass and stiffness for the mode i :
$$ m_{s,i} = \frac{\{U_{0,i} \}^T[M_s]\{U_{0,i} \}}{\{U_{0,i} \}^T\{U_{0,i}\}} \\
k_{s,i} = \frac{\{U_{0,i} \}^T[K_s]\{U_{0,i} \}}{\{U_{0,i} \}^T\{U_{0,i}\}}
$$

### Synthesis of the computation 

Here is the pratical method to calculate the added mass for each modes :

* Solve the natural eigen-value structural problem
* Solve the full system (acoustic/structural) to obtain the eigenvalues $\omega_i$ with the added mass
* Place each modes from the structural system with the modes from the full acoustic/ structural system
* Compute the added mass for each modes by calculating $  \frac{1}{\omega_i ^ 2}k_{s,i} - m_{s,i}$

Now we calculate the right strucutral mode that interests us. We can then extract the frequencies which fit to this strucutral mode and calculate the addded mass, for each density. Effectively, once the right frequency is obtained, the added mass can obtained as follows :
$\omega = 2\pi f$ and $\omega = \sqrt{\frac{k}{m}} $ :
$$ m = \frac{k}{4\pi^2f^2}$$

If we note that $m_0^i$ is the mass of the mode $i$ without fluid, and if we assume that for a given mode $i$, the mode-shapes are constant as the frequency is varying, for the density of fluid $\rho$, the added mass of the mode $i$ verifies:

$$\Delta m^i_\rho = m^i_\rho - m^i_0 = \frac{k}{4\pi^2} \left( \frac{1}{f_{i,\rho}^2} -  \frac{1}{f_{i,0}^2} \right)$$

```python
parametric_result = ERA.load_object(working_folder / "saving_folder/disc_param_study/disc_confined", "dic")
results = parametric_result.results
nb_models=len(results)
print(nb_models)
final_frequencies=[]
added_mass=np.zeros(nb_models-1)
for k in range (0,nb_models):#k=0 corresponds to structural results
    result=results[k]
    frequencies, ids = result.unique_frequencies(decimal = 4, ids_ = True)
    if k==0:#we keep the main structural mode
        pos=result.nodes[0,:]
        print("position initiale :", pos)
        main_struct_frequency,main_id=0,0#future f and id  of the main mode
        main_struct_dpct=result.nodal_results[0,1,0]-pos[1]#we initialise the biggest displacement
        for (frequency, id) in zip(frequencies,ids):
            struct_dpct=result.nodal_results[0,1,id]-pos[1]
            if abs(struct_dpct)>=abs(main_struct_dpct):#we find the main mode by the largest spring elongation
                main_struct_dpct=struct_dpct
                main_struct_frequency,main_id=frequency,id
        print("main structural mode :",main_struct_frequency,main_id)
    else : 
        final_freq,final_id = result.get_frequency(main_struct_frequency, id = True)
        final_frequencies.append(final_freq)
        the_added_mass = 40000/(4*np.pi**2)*(1/final_freq**2-1/main_struct_frequency**2)
        added_mass[k-1]=the_added_mass
print(added_mass)
```

Then we extract the list of the fluid density of the disc parametric study :

Now let's compute $\frac{k}{4\pi^2} \left( \frac{1}{f_{i,\rho}^2} -  \frac{1}{f_{i,0}^2} \right)$ :

```python
#Then we plot the results :
import matplotlib.pyplot as plt
colors =[0,1,2]
plt.plot(density_list[1:], added_mass, linestyle = "none", marker = "+", color = (colors[0], 0, 1 - colors[0]), label = "mode 1")


plt.xlabel("density")
plt.ylabel("added mass")
plt.title("confined disc - numerical solution")


plt.gca().get_yaxis()
#axes = plt.gca()
#axes.set_ylim(-0.01, 0.02)

plt.legend()
```

#### Comparison with the analytical solution

As we learned the method to calculate the added mass of a spring disc system through the speed potentiel $\phi$, we can easily calculate the analytical added mass, knowing the boundary conditions : 

en $r =R_1$ : $u_r(R_1)=Ucos(\theta)$ condition de non pénétration, avec $U$ la vitesse de l'écoulement\
en $r=R_2$ : $u_r(R_2)=0$

On en déduit donc la masse ajoutée $\boxed{m_f=\rho \pi R_1^2\frac{R_2^2+R_1^2}{R_2^2-R_1^2}}$.\
Dans le cas de notre simulation $R_2=2R_1$, nous avons donc : $\boxed{m_f=\frac{5}{3}\rho \pi R_1^2}$


```python
from math import pi
#analytical solution
#print(density_list)
l=len(density_list)
ana_m = np.zeros((1,l))
R=0.2 #try to extract it from the code
for k in range (l):
    rho=density_list[k]
    m=rho*R**2*pi*5/3
    ana_m[0,k]=m
print(ana_m[0])
ana_m=ana_m[0]


%matplotlib notebook
plt.plot(density_list[1:], added_mass,linestyle = "none", marker = "+", color = (colors[0], 0, 1 - colors[0]), label = "numerical")
plt.plot(density_list[:], ana_m, color = (1, 0, 1 - colors[0]), label = "analytical")
plt.title("added mass - confined spring disc")
plt.xlabel("density ($kg/m^3$)")
plt.ylabel("added mass (kg)")


plt.legend()
```

```python
"""
We compute the added mass for the analysis with the fluid
"""
print("--------- Computing of the added mass ---------")
print("-----------------------------------------")
results_fluid = results[1:]
last_freqs = copy(structural_freqs)
added_mass = np.zeros((len(results_fluid) + 1, len(structural_freqs)))
frequency_selected = np.zeros((len(results_fluid) + 1, len(structural_freqs)))
frequency_selected[0,:] = structural_freqs

#To follow the modes evolution during the parametric study :
for idx_param, result_fluid in enumerate(results_fluid):
    approxs_freq = np.zeros((len(structural_freqs),))
    for idx_freq, struc_freq in enumerate(structural_freqs):
        idx_struc_freq = idx_str_freq[idx_freq]
        if idx_param < 2:
                    approx_freq = last_freqs[idx_freq]

        else:
                freq_1 = frequency_selected[idx_param - 1, idx_freq]
                freq_0 = frequency_selected[idx_param - 2, idx_freq]
                approx_freq = 2 * freq_1 - freq_0
        approxs_freq[idx_freq] = approx_freq

    freqs, struct_ids, score = result_fluid.associate_frequencies(approx_freqs=approxs_freq, 
    score = True, family_id=True)
    last_freqs = freqs    
    for idx, freq in enumerate(freqs):
            idx_struc_freq = idx_str_freq[struct_ids[idx]]

            added_mass[idx_param + 1, idx] = \
                    structural_result.compute_added_mass(freq, 
                    idx_struc_freq)
    frequency_selected[idx_param + 1, :] = freqs
print("evolution of the frequencies with the augmentation of the fluid density:")
print(frequency_selected)
print("-----------------------------------------------------------------------")
print("evolution of the added mass with the augmentation of the fluid density :")
print(added_mass)
```

#### Comparison with the analytical expression

To obtain the analytical expression of the added mass by modes, we suppose the n-th mode to have the form in polar coordinates :

$$ w(\theta, t) = q(t) \cos((n + 1) \theta) $$

Here is the analytical expression of the added mass of the problem :

$$ m_a = \pi \frac{\rho}{(n + 1)} R ^ 2$$ 

where $R$ is the radius of the glass, $\rho$ the density of the fluid and $n$ the mode number.

```python
"""
Let's compute the analytical result :
"""
radius = 1
for i in range(5):
    ana_added_mass = (np.pi * radius ** 2 / (i + 3)) * np.array(densitys)
    plt.plot(densitys, ana_added_mass, label = "mode " + str(i + 1))

plt.xlabel("density")
plt.ylabel("added mass")
plt.title("Singing glass - analytical solution")


plt.gca().get_yaxis()
plt.legend()
```

We remarks that there is a proportionality factor between the numeric solution and the analytic solution. Indeed, we don't know how Ansys choose to dimensionate the $M$ and $K$ matrixs. But we can see how the numeric solution evolve with $\rho$. 
With $\rho$ we see that it's the same linear evolution, but how about $n$.

To see how it evolves with $n$ we just take for one specific density the results of all modes, and plot in logarithmic scale.

```python
added_mass_2000 = added_mass[-1,:]
log_mass = np.log(added_mass_2000)
log_n = np.log(np.array([i + 3 for i in range(5)]))
plt.xlabel("log n")
plt.ylabel("log added mass")
plt.title("Singing glass - numerical solution")


plt.gca().get_yaxis()
coefficient = (log_mass[-1] - log_mass[0]) / (log_n[-1] - log_n[0])
print("the power of n : ", coefficient)
plt.plot(log_n, log_mass, linestyle = "none", marker = "+")
plt.gca().get_yaxis()
```

So it's an evolution proportional to $1 / n$ like the analytical solution behaviour.
